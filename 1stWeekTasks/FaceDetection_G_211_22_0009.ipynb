```python
# Instal OpenCV
!pip install opencv-python

import cv2
import matplotlib.pyplot as plt

# Baca gambar yang sudah kamu upload
img = cv2.imread("foto_kelompok.jfif")

# Ubah warna dari BGR ke RGB agar tampil normal
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Load model deteksi wajah
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Konversi gambar ke grayscale untuk deteksi
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Deteksi wajah (nilai 1.1 dan 5 bisa diubah sesuai hasil)
faces = face_cascade.detectMultiScale(gray, 1.1, 5)

print(f"Jumlah wajah terdeteksi: {len(faces)}")

# Data nama dan umur (isi sesuai urutan wajah dari kiri ke kanan)
data_teman = [
    {"nama": "Andi", "umur": 21},
    {"nama": "Arya", "umur": 22},
    {"nama": "Doni", "umur": 20},
    {"nama": "Budi", "umur": 21},
    {"nama": "Syabila", "umur": 23},
    {"nama": "Dhonan", "umur": 22},
    {"nama": "Chika", "umur": 21},
    {"nama": "Revi", "umur": 20},
    {"nama": "Jalu", "umur": 22},
    {"nama": "Cita", "umur": 23},
    {"nama": "Fila", "umur": 22},
    {"nama": "Fatim", "umur": 21},
    {"nama": "Ririn", "umur": 20},
    {"nama": "Agil", "umur": 22},
    {"nama": "Suci", "umur": 23},
    {"nama": "Jors", "umur": 20},
    {"nama": "Fajar", "umur": 22},
    {"nama": "Geo", "umur": 23}
]

# Gambar kotak di wajah dan tulis nama & umur
for i, (x, y, w, h) in enumerate(faces):
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (255, 0, 0), 2)
    if i < len(data_teman):
        teks = f"{data_teman[i]['nama']} ({data_teman[i]['umur']} th)"
        cv2.putText(img_rgb, teks, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

# Tampilkan hasil
plt.figure(figsize=(12,8))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Deteksi Wajah dengan Nama dan Umur")
plt.show()
```

```python
# Install library
!pip install opencv-python matplotlib

# Import library
import cv2
import matplotlib.pyplot as plt

# Upload gambar (3 foto wajah)
from google.colab import files
uploaded = files.upload()  # upload orang1.jpg, orang2.jpg, orang3.jpg

# Data nama dan usia
data_wajah = [
    {"file": "Fila.jpg", "nama": "Fila", "usia": 22}
]

# Load classifier deteksi wajah bawaan OpenCV
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Loop untuk menampilkan hasil deteksi
for data in data_wajah:
    # Baca gambar
    img = cv2.imread(data["file"])
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Deteksi wajah
    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    # Gambar kotak dan teks nama+usia
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img, f"{data['nama']}, {data['usia']} th", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (36,255,12), 2)

    # Tampilkan hasil
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"Wajah: {data['nama']} (Usia: {data['usia']} th)")
    plt.show()
```
